```python
data = pd.read_excel(r'Time series project/Data/ATM/NN5.xlsx')
```

```python
data['Date'] = pd.to_datetime(data['Date'])

# Set 'Date' as the index for easy time series manipulation - for running the code for the first two parts, comment the next line, but for the clustering part leave it as is.
#data.set_index('Date', inplace=True)
data.replace(0, pd.NA, inplace=True)  # Replace zeros with NaN
data.fillna(data.mean(), inplace=True)  # Fill NaN with the column mean
```

```python
import numpy as np
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import time

# Assuming the data is already loaded as 'data'
n_test = 30

# Placeholder for results
results = []

# Loop over each column (excluding the 'Date' column)
for col in data.columns:
    if col == 'Date':
        continue  # Skip the 'Date' column

    print(f'Training Prophet model for column: {col}')

    # Prepare the data for Prophet (requires 'ds' and 'y' columns)
    df = data[['Date', col]].rename(columns={'Date': 'ds', col: 'y'})

    # Split into train and test
    train = df[:-n_test]
    test = df[-n_test:]

    # Initialize Prophet model
    model = Prophet()

    # Measure training time
    start_time = time.time()

    # Fit the model
    model.fit(train)

    # Record the time taken
    training_time = time.time() - start_time

    # Make predictions on the test set
    future = model.make_future_dataframe(periods=n_test, freq='D')
    forecast = model.predict(future)

    # Extract the predictions for the test period
    y_pred = forecast['yhat'][-n_test:].values
    y_test = test['y'].values

    # Calculate MSE and MAPE
    mse = mean_squared_error(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)

    # Store the results
    results.append({
        'Column': col,
        'MSE': mse,
        'MAPE': mape,
        'Training Time (s)': training_time
    })

# Convert the results into a DataFrame and display it
results_df = pd.DataFrame(results)

# Save the DataFrame to a CSV file for download
results_df.to_csv('ATM_Prophet_univariate.csv', index=False)

# Optionally, download the file (if you're on Google Colab or similar)
from google.colab import files
files.download('ATM_Prophet_univariate.csv')
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from tslearn.clustering import TimeSeriesKMeans
from sklearn.preprocessing import StandardScaler

# Assuming the data is already loaded as 'data'
# Exclude the 'Date' column if present, keep only the series columns for clustering
time_series_data = data.iloc[:, 1:].copy()

# Normalize the data for clustering
scaler = StandardScaler()
normalized_data = scaler.fit_transform(time_series_data.T)  # Transpose to have series in rows

# Get the column names (time series identifiers)
column_names = time_series_data.columns.tolist()

# 1. Euclidean Distance Clustering (KMeans)
print("Performing KMeans clustering using Euclidean distance...")

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(normalized_data)
euclidean_labels = kmeans.labels_

# List of column names in each cluster (Euclidean)
euclidean_cluster_1 = [column_names[i] for i in np.where(euclidean_labels == 0)[0].tolist()]
euclidean_cluster_2 = [column_names[i] for i in np.where(euclidean_labels == 1)[0].tolist()]

print(f"Cluster 1 (Euclidean): {euclidean_cluster_1}")
print(f"Cluster 2 (Euclidean): {euclidean_cluster_2}")

# 2. DTW Clustering (TimeSeriesKMeans)
print("Performing DTW-based clustering...")

dtw_kmeans = TimeSeriesKMeans(n_clusters=2, metric="dtw", random_state=42)
dtw_kmeans.fit(time_series_data.T)  # Using transpose to cluster the series
dtw_labels = dtw_kmeans.labels_

# List of column names in each cluster (DTW)
dtw_cluster_1 = [column_names[i] for i in np.where(dtw_labels == 0)[0].tolist()]
dtw_cluster_2 = [column_names[i] for i in np.where(dtw_labels == 1)[0].tolist()]

print(f"Cluster 1 (DTW): {dtw_cluster_1}")
print(f"Cluster 2 (DTW): {dtw_cluster_2}")
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from itertools import product
import random


# Assuming `data` is your main dataset with the Date column
date_column = data[['Date']]  # Extract the Date column

# Define the tuning parameters
changepoint_prior_scales = [0.05, 0.1, 0.5]
seasonality_modes = ['additive', 'multiplicative']
seasonality_prior_scales = [10, 20, 30]

# Ensure Date column is properly formatted in the dataset
def preprocess_dates(df):
    try:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df = df.dropna(subset=['Date'])  # Drop rows where the date is invalid
        df = df.set_index('Date').asfreq('D')  # Ensure daily frequency
    except Exception as e:
        print("Date processing error:", e)
    return df

# Function to train Prophet model with tuning
def train_prophet_with_tuning(df, changepoint_prior_scales, seasonality_modes, seasonality_prior_scales, n_test):
    best_model = None
    best_mse = float('inf')
    best_hyperparams = None
    results = []

    # Iterate over all combinations of hyperparameters
    for cps, seasonality_mode, sps in product(changepoint_prior_scales, seasonality_modes, seasonality_prior_scales):
        print(f"Hyperparameters: Changepoint Prior Scale: {cps}, Seasonality Mode: {seasonality_mode}, Seasonality Prior Scale: {sps}")

        # Configure the Prophet model with current hyperparameters
        model = Prophet(
            changepoint_prior_scale=cps,
            seasonality_mode=seasonality_mode,
            seasonality_prior_scale=sps
        )

        # Fit the model on the training data
        model.fit(df)

        # Make future dataframe and predictions
        future = model.make_future_dataframe(periods=n_test)
        forecast = model.predict(future)

        # Calculate MSE and MAPE for the test set
        y_pred = forecast['yhat'].values[-n_test:]
        y_true = df['y'].values[-n_test:]
        mse = mean_squared_error(y_true, y_pred)
        mape = mean_absolute_percentage_error(y_true, y_pred)

        # Keep track of the best model and hyperparameters
        if mse < best_mse:
            best_mse = mse
            best_model = model
            best_hyperparams = (cps, seasonality_mode, sps)

        # Store results
        results.append({
            'Changepoint Prior Scale': cps,
            'Seasonality Mode': seasonality_mode,
            'Seasonality Prior Scale': sps,
            'MSE': mse,
            'MAPE': mape
        })

    return pd.DataFrame(results), best_model, best_hyperparams

# Function to process and predict for a cluster
# Function to process and predict for a cluster
def process_cluster_with_prophet(cluster_data, cluster_name, n_test=30):
    # Select a random series from the cluster
    selected_series = random.choice(cluster_data.columns)
    print(f"Randomly selected time series for {cluster_name}: {selected_series}")

    # Prepare the data for Prophet (rename columns as required by Prophet)
    series_df = cluster_data[['Date', selected_series]].rename(columns={'Date': 'ds', selected_series: 'y'})
    series_df = preprocess_dates(series_df)

    # Train Prophet model with hyperparameter tuning
    tuning_results, best_model, best_hyperparams = train_prophet_with_tuning(
        series_df, changepoint_prior_scales, seasonality_modes, seasonality_prior_scales, n_test
    )

    # Predict the rest of the series in the cluster using the best model
    results = []
    for col in cluster_data.columns:
        series_to_predict = cluster_data[['Date', col]].rename(columns={'Date': 'ds', col: 'y'})
        series_to_predict = preprocess_dates(series_to_predict)

        # Make future dataframe and predictions
        future = best_model.make_future_dataframe(periods=n_test)
        forecast = best_model.predict(future)

        # Determine the valid test length
        valid_n_test = min(n_test, len(series_to_predict['y']), len(forecast['yhat']))

        # Extract the last valid_n_test samples
        y_pred = forecast['yhat'].values[-valid_n_test:]
        y_true = series_to_predict['y'].values[-valid_n_test:].flatten()

        # Ensure y_true and y_pred lengths match for consistent metrics calculation
        if len(y_true) != len(y_pred):
            print(f"Skipping {col} due to inconsistent lengths: y_true={len(y_true)}, y_pred={len(y_pred)}")
            continue

        # Calculate MSE and MAPE
        mse = mean_squared_error(y_true, y_pred)
        mape = mean_absolute_percentage_error(y_true, y_pred)

        # Store results
        results.append({
            'Cluster': cluster_name,
            'Time Series': col,
            'Selected Series for Training': selected_series,
            'MSE': mse,
            'MAPE': mape,
            'Best Hyperparameters': best_hyperparams
        })

    return pd.DataFrame(results)




# Assuming you have already created clusters (euclidean_cluster_1, euclidean_cluster_2, dtw_cluster_1, dtw_cluster_2)
# Process each cluster and train a Prophet model for one series in each

# Define each cluster with Date appended only when calling the function
euclidean_cluster_1_results = process_cluster_with_prophet(pd.concat([data[['Date']], data[euclidean_cluster_1]], axis=1), "Euclidean Cluster 1")
euclidean_cluster_2_results = process_cluster_with_prophet(pd.concat([data[['Date']], data[euclidean_cluster_2]], axis=1), "Euclidean Cluster 2")
dtw_cluster_1_results = process_cluster_with_prophet(pd.concat([data[['Date']], data[dtw_cluster_1]], axis=1), "DTW Cluster 1")
dtw_cluster_2_results = process_cluster_with_prophet(pd.concat([data[['Date']], data[dtw_cluster_2]], axis=1), "DTW Cluster 2")


# Combine all results into one DataFrame
all_results = pd.concat([euclidean_cluster_1_results, euclidean_cluster_2_results, dtw_cluster_1_results, dtw_cluster_2_results])

# Save the combined results to a CSV file (optional)
# all_results.to_csv('clustered_prophet_results.csv', index=False)

# Show the results
#print(all_results)
```

# multivariate prophet

```python
from prophet import Prophet
import numpy as np
import pandas as pd
import random
from sklearn.preprocessing import MinMaxScaler

# Scaling function for multivariate data
def scale_multivariate_data(data):
    scaler = MinMaxScaler()
    return scaler, scaler.fit_transform(data)

# Function to process a cluster with Prophet model tuning on a single series
def process_cluster_with_tuning(cluster_data, cluster_name, date_series):
    results = []
    # Define parameter grid for tuning
    param_grid = {
        'changepoint_prior_scale': [0.05, 0.1, 0.5],
        'seasonality_mode': ['additive', 'multiplicative'],
        'seasonality_prior_scale': [10, 20, 30]
    }

    # Randomly select a single target series for tuning
    selected_series = random.choice(cluster_data.columns.drop('Date'))
    print(f"Tuning model on randomly selected series '{selected_series}' in cluster '{cluster_name}'")

    # Prepare the dataset with date and selected series as 'y'
    df_selected = date_series.copy()
    df_selected['y'] = cluster_data[selected_series]

    # Add remaining series as regressors for multivariate Prophet
    regressor_cols = [col for col in cluster_data.columns.drop('Date') if col != selected_series]
    for col in regressor_cols:
        df_selected[col] = cluster_data[col]

    # Track the best model based on MSE
    best_mse = float("inf")
    best_params = None
    best_model = None

    # Tuning loop
    for changepoint_prior_scale in param_grid['changepoint_prior_scale']:
        for seasonality_mode in param_grid['seasonality_mode']:
            for seasonality_prior_scale in param_grid['seasonality_prior_scale']:
                # Initialize Prophet model with parameters
                model = Prophet(
                    changepoint_prior_scale=changepoint_prior_scale,
                    seasonality_mode=seasonality_mode,
                    seasonality_prior_scale=seasonality_prior_scale
                )

                # Add regressors to the model
                for regressor in regressor_cols:
                    model.add_regressor(regressor)

                # Train model and measure performance
                try:
                    model.fit(df_selected)
                    future = model.make_future_dataframe(periods=30)

                    # Include only the last 30 rows of regressors for future data
                    for regressor in regressor_cols:
                        future[regressor] = pd.concat([df_selected[regressor], cluster_data[regressor].tail(30).reset_index(drop=True)], ignore_index=True)

                    forecast = model.predict(future)

                    # Calculate MSE on training data, aligned to the original data length
                    mse = np.mean((forecast['yhat'][:len(df_selected)] - df_selected['y']) ** 2)

                    if mse < best_mse:
                        best_mse = mse
                        best_params = (changepoint_prior_scale, seasonality_mode, seasonality_prior_scale)
                        best_model = model
                except Exception as e:
                    print(f"Error training model with parameters ({changepoint_prior_scale}, {seasonality_mode}, {seasonality_prior_scale}) on '{selected_series}': {e}")
                    continue

    if best_model is None:
        print(f"No valid model found for series '{selected_series}' in cluster '{cluster_name}'. Skipping forecasting for this cluster.")
        return pd.DataFrame([])

    print(f"Best model parameters for '{selected_series}' in cluster '{cluster_name}': {best_params}, MSE: {best_mse}")

    # Apply the best model to all series in the cluster
    for target_series in cluster_data.columns.drop('Date'):
        print(f"Forecasting for series '{target_series}' using tuned model in cluster '{cluster_name}'")

        df_target = date_series.copy()
        df_target['y'] = cluster_data[target_series]

        # Include regressors for multivariate forecast
        for col in regressor_cols:
            df_target[col] = cluster_data[col]

        try:
            future = best_model.make_future_dataframe(periods=30)
            for regressor in regressor_cols:
                # Only append the last 30 rows for the future period
                future[regressor] = pd.concat([df_target[regressor], cluster_data[regressor].tail(30).reset_index(drop=True)], ignore_index=True)

            forecast = best_model.predict(future)

            # Align forecast length with the original data length
            yhat_trimmed = forecast['yhat'][:len(df_target)]
            mse = np.mean((yhat_trimmed - df_target['y']) ** 2)
            mape = np.mean(np.abs((df_target['y'] - yhat_trimmed) / df_target['y'])) * 100

            # Append the results
            results.append({
                'Cluster': cluster_name,
                'Target Series': target_series,
                'MSE': mse,
                'MAPE': mape,
                'Best Parameters': best_params
            })
        except Exception as e:
            print(f"Error forecasting for {target_series}: {e}")
            results.append({
                'Cluster': cluster_name,
                'Target Series': target_series,
                'MSE': None,
                'MAPE': None,
                'Best Parameters': best_params
            })

    return pd.DataFrame(results)

# Assuming date_series and main data are available
date_series = data[['Date']].rename(columns={'Date': 'ds'})

# Define clusters with the appropriate column names for each cluster
euclidean_cluster_1_data = data[['Date'] + euclidean_cluster_1]
euclidean_cluster_2_data = data[['Date'] + euclidean_cluster_2]
dtw_cluster_1_data = data[['Date'] + dtw_cluster_1]
dtw_cluster_2_data = data[['Date'] + dtw_cluster_2]

# Run the processing function on each cluster
euclidean_cluster_1_results = process_cluster_with_tuning(euclidean_cluster_1_data, "Euclidean Cluster 1", date_series)
euclidean_cluster_2_results = process_cluster_with_tuning(euclidean_cluster_2_data, "Euclidean Cluster 2", date_series)
dtw_cluster_1_results = process_cluster_with_tuning(dtw_cluster_1_data, "DTW Cluster 1", date_series)
dtw_cluster_2_results = process_cluster_with_tuning(dtw_cluster_2_data, "DTW Cluster 2", date_series)

# Combine all results into one DataFrame
all_results = pd.concat([euclidean_cluster_1_results, euclidean_cluster_2_results, dtw_cluster_1_results, dtw_cluster_2_results])

# Display or save the results
print(all_results)
```

```python
from prophet import Prophet
import numpy as np
import pandas as pd

# Function to create an averaged series and then apply the tuned model across the cluster
def process_cluster_with_averaged_tuning(data, cluster, cluster_name, date_series):
    results = []

    # Select the columns for the current cluster, including the 'Date' column
    cluster_data = data[['Date'] + cluster]

    # Define parameter grid for tuning
    param_grid = {
        'changepoint_prior_scale': [0.05, 0.1, 0.5],
        'seasonality_mode': ['additive', 'multiplicative'],
        'seasonality_prior_scale': [10, 20, 30]
    }

    # Create an averaged series for tuning
    averaged_series = cluster_data.drop(columns='Date').mean(axis=1)
    df_averaged = date_series.copy()
    df_averaged['y'] = averaged_series

    # Track the best model based on MSE
    best_mse = float("inf")
    best_params = None
    best_model = None

    # Tuning loop using averaged series
    for changepoint_prior_scale in param_grid['changepoint_prior_scale']:
        for seasonality_mode in param_grid['seasonality_mode']:
            for seasonality_prior_scale in param_grid['seasonality_prior_scale']:
                # Initialize Prophet model with parameters
                model = Prophet(
                    changepoint_prior_scale=changepoint_prior_scale,
                    seasonality_mode=seasonality_mode,
                    seasonality_prior_scale=seasonality_prior_scale
                )

                # Train model and measure performance
                try:
                    model.fit(df_averaged)
                    future = model.make_future_dataframe(periods=30)
                    forecast = model.predict(future)

                    # Calculate MSE on training data, aligned to the original data length
                    mse = np.mean((forecast['yhat'][:len(df_averaged)] - df_averaged['y']) ** 2)

                    if mse < best_mse:
                        best_mse = mse
                        best_params = (changepoint_prior_scale, seasonality_mode, seasonality_prior_scale)
                        best_model = model
                except Exception as e:
                    print(f"Error training model with parameters ({changepoint_prior_scale}, {seasonality_mode}, {seasonality_prior_scale}) on averaged series in '{cluster_name}': {e}")
                    continue

    if best_model is None:
        print(f"No valid model found for cluster '{cluster_name}'. Skipping forecasting for this cluster.")
        return pd.DataFrame([])

    print(f"Best model parameters for cluster '{cluster_name}': {best_params}, MSE: {best_mse}")

    # Apply the best model to all series in the cluster
    for target_series in cluster:
        print(f"Forecasting for series '{target_series}' using tuned model in cluster '{cluster_name}'")

        df_target = date_series.copy()
        df_target['y'] = data[target_series]

        try:
            future = best_model.make_future_dataframe(periods=30)
            forecast = best_model.predict(future)

            # Align forecast length with the original data length
            yhat_trimmed = forecast['yhat'][:len(df_target)]
            mse = np.mean((yhat_trimmed - df_target['y']) ** 2)
            mape = np.mean(np.abs((df_target['y'] - yhat_trimmed) / df_target['y'])) * 100

            # Append the results
            results.append({
                'Cluster': cluster_name,
                'Target Series': target_series,
                'MSE': mse,
                'MAPE': mape,
                'Best Parameters': best_params
            })
        except Exception as e:
            print(f"Error forecasting for {target_series}: {e}")
            results.append({
                'Cluster': cluster_name,
                'Target Series': target_series,
                'MSE': None,
                'MAPE': None,
                'Best Parameters': best_params
            })

    return pd.DataFrame(results)

# Assuming date_series and data are available and contain the 'Date' column and all time series columns
date_series = data[['Date']].rename(columns={'Date': 'ds'})

# Define clusters with the appropriate column names for each cluster
# Example: euclidean_cluster_1 = ['NN5 - 001', 'NN5 - 002', ...]
euclidean_cluster_1_data = process_cluster_with_averaged_tuning(data, euclidean_cluster_1, "Euclidean Cluster 1", date_series)
euclidean_cluster_2_data = process_cluster_with_averaged_tuning(data, euclidean_cluster_2, "Euclidean Cluster 2", date_series)
dtw_cluster_1_data = process_cluster_with_averaged_tuning(data, dtw_cluster_1, "DTW Cluster 1", date_series)
dtw_cluster_2_data = process_cluster_with_averaged_tuning(data, dtw_cluster_2, "DTW Cluster 2", date_series)

# Combine all results into one DataFrame
all_results = pd.concat([euclidean_cluster_1_data, euclidean_cluster_2_data, dtw_cluster_1_data, dtw_cluster_2_data])

# Display or save the results
print(all_results)
```
